   C. 矩阵对向量的点乘满足结合律，即对于维度合适的矩阵$A$和向量$\bm{x,y}$有$\bm{(Ax)\cdot y}= \bm{A(x\cdot y)}$。  
   D. 假设$f$处处可微且存在最大值，那么在最大值点$f$的梯度一定为零。  


答：  
A. 正确。  
对角矩阵相乘就是对角元素分别作用于另一矩阵的每一行，由于对角矩阵只有对角线元素，交换顺序并不会影响乘法，因此满足乘法的交换律。  

```python
import numpy as np

# 定义两个对角矩阵D1和D2
D1 = np.diag([1, 2, 3])
D2 = np.diag([4, 5, 6])

# 计算D1和D2的乘积，先乘D1再乘D2
product1 = np.dot(D1, D2)

# 计算D2和D1的乘积，先乘D2再乘D1
product2 = np.dot(D2, D1)

# 输出结果
print("D1 * D2:\n", product1)
print("D2 * D1:\n", product2)

# 检查两个乘积是否相等
print("Are D1 * D2 and D2 * D1 equal? ", np.array_equal(product1, product2))
```

3. 当$L_p$范数中的$p\in (0,1)$或取$p\rightarrow 0^+$得到$L_p$范数，它并不满足范数的定义。依次验证范数的三条要求，它违反了哪一条要求？试举出一个违反该要求的例子。

答：  
$p\in (1,0)$违反了三角不等式

```python
import numpy as np

# 定义两个向量x和y
x = np.array([1, 0, 0])
y = np.array([0, 1, 0])

# 计算x和y的Lp范数，其中p接近于0
p = 0.5
norm_x = np.sum(np.abs(x)**p)**(1/p)
norm_y = np.sum(np.abs(y)**p)**(1/p)

# 计算x+y的Lp范数
norm_z = np.sum(np.abs(x + y)**p)**(1/p)

# 输出结果
print("Lp norm of x:", norm_x)
print("Lp norm of y:", norm_y)
print("Lp norm of x + y:", norm_z)
```

$p\rightarrow 0^+$得到$L_p$范数，不满足三角不等式。
当$p$趋近于0，范数趋向于选择向量中最大的绝对值作为整个向量的范数，而不是按照范数的规则计算。因为这个时候$|x_i|^p$都将趋近于1，最大值将主导整个求和。（虽然都变得微不足道）

```python
import numpy as np
a = 5
b = 500
p = 0.005
q1 = a ** p
q2 = b ** p
print(q1)
print(q2)
```

```python
import numpy as np

x = np.array([1, 0, 10])
y = np.array([1, 20, 4])
p = 0.5

def Lp_norm(p, vector):
    powered_sum = np.sum(np.abs(vector) ** p)
    return powered_sum ** (1 / p)


norm_x = Lp_norm(p, x)
norm_y = Lp_norm(p, y)
norm_xy = Lp_norm(p, x + y)

print("Lp norm of x:", norm_x)
print("Lp norm of y:", norm_y)
print("Lp norm of x + y:", norm_xy)

if norm_xy <= norm_x + norm_y:
    print('True')
else:
    print('False')
```

4. 证明矩阵的转置和逆满足$(A^T)^{-1}=(A^{-1})^T$.

设矩阵$A$可逆  
$AA^{-1}=I$  
$(AA^{-1})^T=(A^{-1})^TA^T=I$  
$A^T(A^T)^{-1}=I$  
故$(A^T)^{-1}=(A^{-1})^T$

```python
import numpy as np

# 设置随机种子以获得可重复的结果
np.random.seed(0)

# 生成一个随机的可逆矩阵A
n = 4  # 选择矩阵的大小
A = np.random.rand(n, n)
A = np.dot(A, A.T)  # 确保A是可逆的

# 计算矩阵A的逆矩阵
A_inv = np.linalg.inv(A)

# 计算矩阵A的转置
A_T = A.T

# 计算转置矩阵的逆矩阵
A_T_inv = np.linalg.inv(A_T)

# 计算逆矩阵的转置
A_inv_T = A_inv.T

print(A_T_inv)
print(A_inv_T)
if np.allclose(A_inv_T, A_T_inv): # allclose能解决浮点的误差小问题
    print('True')
else:
    print('False')
```

5.设二元函数$f(x,y)$在$\mathbb{R}^2$上处处光滑且可微，证明在任意一点$(x_0, y_0)$处，函数的梯度是函数值上升最快的方向。（提示：考虑函数在该店沿$(\cos\theta,\sin\theta)$的方向导数的长度，何时长度最大？）  
ps:这个问题对理解梯度下降非常重要

设二元函数 $ f(x, y) $ 在 $ \mathbb{R}^2 $ 上处处光滑且可微，我们想要证明在任意一点 $ (x_0, y_0) $ 处，函数的梯度 $ \nabla f(x_0, y_0) $ 指向函数值上升最快的方向。

梯度 $ \nabla f(x_0, y_0) $ 定义为：

$ \nabla f(x_0, y_0) = \left( \frac{\partial f}{\partial x}(x_0, y_0), \frac{\partial f}{\partial y}(x_0, y_0) \right) $

现在，考虑在点 $ (x_0, y_0) $ 处，函数 $ f $ 沿任意单位向量 $ \mathbf{u} = (\cos\theta, \sin\theta) $ 的方向导数。根据方向导数的定义，我们有：

$ D_{\mathbf{u}}f(x_0, y_0) = \nabla f(x_0, y_0) \cdot \mathbf{u} $

点乘$ \cdot $ 表示向量的点乘。计算得到：

$ D_{\mathbf{u}}f(x_0, y_0) = \left( \frac{\partial f}{\partial x}(x_0, y_0) \cos\theta + \frac{\partial f}{\partial y}(x_0, y_0) \sin\theta \right) $

我们知道，向量 $ \mathbf{u} $ 的长度是 1，而梯度 $ \nabla f(x_0, y_0) $ 的长度是 $ |\nabla f(x_0, y_0)| $。因此，方向导数的长度（即方向导数的绝对值）可以表示为：

$ |D_{\mathbf{u}}f(x_0, y_0)| = \left| \nabla f(x_0, y_0) \right| \cos(\theta) $

为了找到使得 $ |D_{\mathbf{u}}f(x_0, y_0)| $ 最大的方向，我们需要找到使得 $ \cos(\theta) $ 最大的 $ \theta $。显然，当 $ \cos(\theta) = 1 $ 时，即 $ \theta = 0 $ 时，$ \cos(\theta) $ 取得最大值 1。

因此，方向导数的长度 $|D_{\mathbf{u}}f(x_0, y_0)|$ 在 $ \theta = 0  $时最大，此时 $ \mathbf{u} $ 与梯度 $ \nabla f(x_0, y_0) $ 的方向相同。这意味着函数 $ f $ 在 $ (x_0, y_0) $ 处沿梯度方向 $ \nabla f(x_0, y_0) $ 的方向导数的长度最大，即梯度方向是函数值上升最快的方向。


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 定义一个二元函数，例如 f(x, y) = x^2 + y^2
def f(x, y):
    return x**2 + y**2

# 选择一个特定的点 (x0, y0) 来计算梯度
x0, y0 = 0, 0  # 选择原点

# 创建一个网格来计算梯度
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 计算梯度的 x 和 y 分量
grad_x, grad_y = np.gradient(Z, x, y)

# 创建3D图形
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制函数的表面
ax.plot_surface(X, Y, Z, color='b', alpha=0.5)

# 绘制梯度方向
# 绘制一条从点 (x0, y0) 出发，沿着梯度方向的线
# 注意：quiver需要U, V, W (x, y, z速度分量)，这里Z方向为0
ax.quiver(x0, y0, 0, grad_x[x0, y0], grad_y[x0, y0], 0, color='r', arrow_length_ratio=0.1)

# 设置图形的标签和范围
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')

# 显示图形
plt.show()
```

6. 利用向量范数的定义证明，所以的向量范数都是凸函数。

向量范数是一个函数 $ \|\cdot\|: \mathbb{R}^n \rightarrow [0, \infty) $，它将每个向量映射到一个非负实数。一个函数 $ f $ 被称为凸函数，如果对于所有 $ \mathbf{x}, \mathbf{y} \in \mathbb{R}^n $ 和所有 $ \lambda \in [0, 1] $，都有：

$ f(\lambda \mathbf{x} + (1 - \lambda) \mathbf{y}) \leq \lambda f(\mathbf{x}) + (1 - \lambda) f(\mathbf{y}) $

为了证明所有的向量范数都是凸函数，我们需要证明对于任意两个向量 $ \mathbf{x} $ 和 $ \mathbf{y} $ 以及任意的 $ \lambda \in [0, 1] $，范数满足上述不等式。

设 $ \|\cdot\| $ 是一个向量范数，那么根据范数的定义，存在一个仿射函数 $ h: \mathbb{R}^n \rightarrow \mathbb{R} $ 使得 $ \|\mathbf{v}\| = h(\mathbf{v}) $ 对于所有 $ \mathbf{v} $ 都成立，并且 $ h(\mathbf{v}) $ 是关于 $ \mathbf{v} $ 的凸函数。

我们可以选择 $ h(\mathbf{v}) = \sup_{\|\mathbf{u}\| = 1} \{\mathbf{v} \cdot \mathbf{u} \} $，这是一个关于 $ \mathbf{v} $ 的凸函数，因为它是 $ \mathbf{v} $ 和单位球面上的点 $ \mathbf{u} $ 的点乘的最大值。

现在，让我们证明范数函数 $ \|\cdot\| $ 是凸的。对于任意的 $ \mathbf{x}, \mathbf{y} \in \mathbb{R}^n $ 和 $ \lambda \in [0, 1] $，我们有：

$ \|\lambda \mathbf{x} + (1 - \lambda) \mathbf{y}\| = h(\lambda \mathbf{x} + (1 - \lambda) \mathbf{y}) $

由于 $ h $ 是凸函数，我们可以得到：

$ h(\lambda \mathbf{x} + (1 - \lambda) \mathbf{y}) \leq \lambda h(\mathbf{x}) + (1 - \lambda) h(\mathbf{y}) $

这等同于：

$ \|\lambda \mathbf{x} + (1 - \lambda) \mathbf{y}\| \leq \lambda \|\mathbf{x}\| + (1 - \lambda) \|\mathbf{y}\| $

这就证明了所有的向量范数都是凸函数。这个性质在优化理论和数值分析中非常重要，因为它意味着最小化范数（例如，在求解最小二乘问题时）可以通过凸优化技术来有效解决。

7. 是否存在非凸非凹的函数？又凸又凹呢？举例说明或证明其不存在。

在数学中，一个函数在某区间上被称为凸函数，如果对于该区间内的任意两点和任意实数 $ \lambda $ 满足 $ 0 \leq \lambda \leq 1 $，都有：

$ f(\lambda \mathbf{x} + (1 - \lambda) \mathbf{y}) \leq \lambda f(\mathbf{x}) + (1 - \lambda) f(\mathbf{y}) $

类似地，一个函数在某区间上被称为凹函数，如果对于该区间内的任意两点和任意实数 $ \lambda $ 满足 $ 0 \leq \lambda \leq 1 $，都有：

$ f(\lambda \mathbf{x} + (1 - \lambda) \mathbf{y}) \geq \lambda f(\mathbf{x}) + (1 - \lambda) f(\mathbf{y}) $

对于一个函数来说，它在某区间上不可能同时是凸的和凹的，因为这将意味着对于区间内的任意两点，函数值在这两点的线性组合处既大于又小于这两点处函数值的线性组合，这显然是矛盾的。

然而，可以存在一个函数，在某个区间上是凸的，在另一个区间上是凹的。例如，考虑函数：

$ f(x) = -x^2 $

这个函数在区间 $ (-\infty, 0] $ 上是凹的，因为在该区间内，对于任意的 $ x_1, x_2 $ 和 $ \lambda $，都有：

$ f(\lambda x_1 + (1 - \lambda) x_2) = -(\lambda x_1 + (1 - \lambda) x_2)^2 \geq \lambda (-x_1^2) + (1 - \lambda) (-x_2^2) $

但是，这个函数在区间 $ [0, \infty) $ 上是凸的，因为：

$ f(\lambda x_1 + (1 - \lambda) x_2) = -(\lambda x_1 + (1 - \lambda) x_2)^2 \leq \lambda (-x_1^2) + (1 - \lambda) (-x_2^2) $

所以，函数 $ f(x) = -x^2 $ 在 $ (-\infty, 0] $ 上是凹的，在 $ [0, \infty) $ 上是凸的。

至于是否存在非凸非凹的函数，答案是肯定的。例如，考虑绝对值函数：

$ g(x) = |x| $

这个函数在 $ (-\infty, 0) $ 和 $ (0, \infty) $ 上是凸的，但在 $ (-\infty, 0) $ 和 $ (0, \infty) $ 之间的区间 $ \{0\} $ 上既不是凸的也不是凹的。在这个例子中，函数在 $ x = 0 $ 处不是凸的，因为对于 $ x_1 = -\epsilon $ 和 $ x_2 = \epsilon $（其中 $ \epsilon > 0 $），以及 $ \lambda = 0.5 $，我们有：

$ g(\lambda x_1 + (1 - \lambda) x_2) = g(0) = 0 $

而

$ \lambda g(x_1) + (1 - \lambda) g(x_2) = \lambda |-\epsilon| + (1 - \lambda) |\epsilon| = 2\epsilon \neq 0 $

所以，绝对值函数 $ g(x) = |x| $ 在 $ x = 0 $ 处不是凸的，也不是凹的。这表明函数可以在某些区间上是非凸非凹的。

8. 试通过作图来展示，对不同的$p$值，在2维坐标上画出$L_p$范数等于1的向量对应的点构成的曲线。设平面上点的坐标为$(x,y)$，该去点的方程就是$(|x|^p+|y|^p)^{1/p}=1$。通过这些图像来证明和理解一个趋势：$p$越小，曲线越贴近坐标轴；$p$越大，曲线越远离坐标轴，并且棱角越明显。

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 设置 $ p $ 的值
ps = [0.5, 1, 1.5, 2]

# 定义一个函数来计算 $ L_p $ 范数等于 1 的点
def Lp_norm_points(p):
    # 生成极坐标点
    theta = np.linspace(0, 2 * np.pi, 1000)
    r = 1 / (np.abs(np.cos(theta))**(1/p) + np.abs(np.sin(theta))**(1/p))
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return x, y

# 绘制 $ L_p $ 范数曲线
fig, ax = plt.subplots()
for p in ps:
    x, y = Lp_norm_points(p)
    # 绘制曲线
    curve = patches.Polygon(np.column_stack((x, y)), closed=True, fill=False, edgecolor='blue')
    ax.add_patch(curve)

    # 标注 $ p $ 值
    ax.text(0.9 * np.mean(x), 0.9 * np.mean(y), f"$p={p:.1f}$", fontsize=12, color='red')

# 设置坐标轴
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_aspect('equal')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title('$L_p$范数曲线')

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义范围和密度
x = np.linspace(-1.5, 1.5, 400)
y = np.linspace(-1.5, 1.5, 400)

# 创建网格
X, Y = np.meshgrid(x, y)

# 不同的 p 值
p_values = [0.1, 0.5, 1, 2, 5, 10]

plt.figure(figsize=(12, 8))

for p in p_values:
    # 计算方程左侧的值
    left_side = np.abs(X)**p + np.abs(Y)**p

    # 计算曲线上的点
    X_curve = X[left_side**(1/p) == 1]
    Y_curve = Y[left_side**(1/p) == 1]

    # 绘制曲线
    plt.plot(X_curve, Y_curve, label=f'p={p}')

plt.title('Curves of $L_p$ Norm Equals 1 for Different $p$ Values')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```
