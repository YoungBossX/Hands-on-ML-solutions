1. 以下关于最大似然估计的表述中正确的是：   
   A. 以概率为输出的模型常用最大似然估计得到损失函数。

   C. 最大似然估计与交叉熵的训练目标不等价。   
   D. 最大似然估计中引入了概率分布，所以不能采用梯度下降法来优化最大似然估计导出的损失函数。

B. 错误。最大似然估计优化的是似然函数，实践中由于计算和优化问题，会将似然值取对数优化计算。

2. 以下关于分类问题的说法中不正确的是：   
   A. 分类问题中，最后往往需要通过阈值来决定样本最后的标签。对于标签为 0 或 1 的二分类问题，当$f_{\theta}(x)$的数值大于 0.5 时即可认为标签为 1，反之亦然。   
   B. 如果使用确定性模型，将会导致模型对于参数无法微分，因此我们需要使用概率模型来建模问题。   
   C. 对于多分类问题，在设计损失函数时仍然可以采用交叉熵损失。如果有$k$个类别，损失函数就是每一类的损失相乘。   
   D. softmax 函数可以看作是逻辑斯谛函数在多分类情况下的延伸，因此也可以用 softmax 函数作为二分类问题的损失函数。  

答：  
A. 正确。分类问题中，会使用激活函数将模型的输出转化为概率，然后通过阈值来决定样本的标签。这样才能实现预测样本属于某个类别，如果输出的是确定值，则不知道怎么进行划分。    
B. 正确。分类问题不同于回归问题输出是确定值的模型，分类问题需要根据概率来进行分类，如果输出的不是概率，无法使用损失函数来梯度下降优化。分类问题中的输出不像回归问题中的是连续值，分类问题输出的是预测的类别的概率才能使用梯度下降法进行优化，根据预测概率和真实标签来计算损失函数的梯度。  
C. 错误。多分类问题使用交叉熵是每一个类别的损失相加而不是相乘。因为分类模型损失函数的值是一非负数，太多类别相乘的话会使得损失函数的值变得很小，梯度下降会很慢和很难收敛。  
D. 正确。softmax函数是一种激活函数，是将输出转化为概率，和逻辑斯蒂回归一样都使得输出的值在0到1之间。不过softmax还要求所有分量的和为1。逻辑斯蒂回归只有一个输出，softmax函数有多个输出，是输出一个类别，就相当于是逻辑斯蒂回归在多分类回归下的延申。在二分类问题中，softmax函数将简化为sigmoid函数，即为$k=0,p(y=1|x)=\frac{1}{1+e^{-f(x)}}$，然后使用二元交叉熵（最大似然估计）作为损失函数训练模型

答:逻辑斯谛回归的参数化假设是线性的。  
$p$为样本为正例的概率，那么模型假设$p$与样本特征$x$之间存在线性关系：  
$\log (\frac{p}{1-p})=w^Tx+b$  
逻辑斯谛回归模型是引入非线性，但是参数可以通过线性代数进行估计，引入非线性只是为了将输出转化为概率。

6. 对于一个二分类任务，数据的标签和对于预测正例的概率如下表所示，试画出 ROC 曲线并计算模型的 AUC 值。  
表6-4习题6概率表
    |  $n_1$  |  $n_2$  |  $n_3$  | $n_4$ | $p_1$ | $p_2$ | $p_3$ | $p_4$ |
    |  :----:  | :----:  | :----:  | :----:  | :----:  | :----:  | :----:  | :----:  |
    |  0.15  |  0.21  |  0.74  | 0.45  |  0.71  |  0.48  |  0.52  |  0.34  |
  
ps：n负类，p正类，数字预测为正例的概率  
画出来明显看到这个分类的结果是不好的

```python
import numpy as np
import matplotlib.pyplot as plt

y_true = np.array([1, 1, 1, 1, 0, 0, 0, 0]) # 0负，1正
y_pred = np.array([0.15, 0.21, 0.74, 0.45, 0.71, 0.48, 0.52, 0.34])
idx = np.argsort(y_pred)[::-1]
y_true = y_true[idx]
y_pred = y_pred[idx]

tpr_list = []
fpr_list = []
thresholds = np.linspace(0, 1, 100) # 阈值
for threshold in thresholds:
    predicted_labels = (y_pred >= threshold).astype(int)

    TP = np.sum((predicted_labels == 1) & (y_true == 1))
    FP = np.sum((predicted_labels == 1) & (y_true == 0))
    TN = np.sum((predicted_labels == 0) & (y_true == 0))
    FN = np.sum((predicted_labels == 0) & (y_true == 1))

    TPR = TP / (TP + FN)
    FPR = FP / (FP + TN)

    tpr_list.append(TPR)
    fpr_list.append(FPR)

# 计算 AUC
auc = np.trapz(tpr_list[::-1], fpr_list[::-1])

plt.figure(figsize=(10, 8))
plt.plot(fpr_list, tpr_list, color='darkorange', lw=2)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--') # 修改这里，绘制45度对角线
plt.fill_between(fpr_list, tpr_list, color='lightblue', alpha=0.5)  # 给 AUC 区域上色
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC')
plt.show()

print(f"AUC:{auc}")
```

8. 对于$K$分类 softmax 函数，试推导其中一个分类的逻辑斯谛值总可以设置为0，进而$K$分类逻辑斯谛回归模型其实只需要使用$K-1$个参数向量即可完成等价建模，而具体的二分类逻辑斯谛回归的形式则正是 softmax 函数在$K=2$时由这样简化得到的。

答：  
$K$ 分类 softmax 函数的定义。 softmax 函数将 $K$ 个实数映射到 $[0, 1]$ 区间，使得这 $K$ 个数的和为 $1$，因此可以用来表示 $K$ 个类别的概率分布。具体来说，对于给定的输入 $x$，softmax 函数的输出为：
$$
\text{softmax}(x)_i = \frac{e^{x_i}}{\sum_{j=1}^{K} e^{x_j}}, \quad i = 1, 2, \ldots, K.
$$
其中 $x_i$ 表示第 $i$ 个类别的得分，$e^{x_i}$ 表示第 $i$ 个类别的非归一化概率，$\sum_{j=1}^{K} e^{x_j}$ 表示所有类别的非归一化概率之和。

现在，我们来证明对于 $K$ 分类 softmax 函数，其中一个分类的逻辑斯谛值总可以设置为 $0$。我们假设第 $K$ 个类别的逻辑斯谛值为 $0$，那么 softmax 函数可以写成：

$$
\begin{aligned}
\text{softmax}(x)_i &= \frac{e^{x_i}}{e^{x_K} + \sum_{j=1}^{K-1} e^{x_j}} \\
&= \frac{1}{1 + \sum_{j=1}^{K-1} e^{x_j - x_K}}, \quad i = 1, 2, \ldots, K-1, \\
\text{softmax}(x)_K &= \frac{e^{x_K}}{e^{x_K} + \sum_{j=1}^{K-1} e^{x_j}} \\
&= \frac{1}{1 + \sum_{j=1}^{K-1} e^{x_j - x_K}}.
\end{aligned}
$$

可以看出，当第 $K$ 个类别的逻辑斯谛值为 $0$ 时，softmax 函数实际上只依赖于前 $K-1$ 个类别的逻辑斯谛值。因此，我们可以使用 $K-1$ 个参数向量来表示 $K$ 分类逻辑斯谛回归模型。

接下来，我们来看看当 $K=2$ 时，softmax 函数会简化成什么形式。当 $K=2$ 时，softmax 函数可以写成：

$$
\begin{aligned}
\text{softmax}(x)_1 &= \frac{e^{x_1}}{e^{x_1} + e^{x_2}} \\
&= \frac{1}{1 + e^{-(x_1 - x_2)}}, \\
\text{softmax}(x)_2 &= \frac{e^{x_2}}{e^{x_1} + e^{x_2}} \\
&= \frac{1}{1 + e^{x_1 - x_2}}.
\end{aligned}
$$

可以看出，当 $K=2$ 时，softmax 函数实际上就是二分类逻辑斯谛回归模型的形式。因此，我们可以将 $K$ 分类逻辑斯谛回归模型看作是 $K-1$ 个二分类逻辑斯谛回归模型的组合，每个二分类模型负责判断一个类别与其他所有类别的概率。

因此，我们证明了对于 $K$ 分类 softmax 函数，其中一个分类的逻辑斯谛值总可以设置为 $0$，进而 $K$ 分类逻辑斯谛回归模型其实只需要使用 $K-1$ 个参数向量即可完成等价建模，而具体的二分类逻辑斯谛回归的形式则正是 softmax 函数在 $K=2$ 时由这样简化得到的。

9. 从二分类到多分类  
    (1) 推导 softmax 函数$\sigma (z)$对$z$的梯度。  
      
    (2) 将$\sigma (z)$作为模型预测的概率分布，分别用 MLE 和交叉熵计算$K$分类问题的损失函数。两者的结果是否相同？  
      
    (3) 利用(1)和(2)的结果实现多分类的梯度下降算法，并在`multiclass.csv`数据集上测试。该数据集每一行包括三个数字，依次为样本的$x$坐标、$y$坐标和标签。数据分布如图6-8所示。!["图6-8多分类数据集的数据分布"](https://hml.boyuai.com/static/multiclass.2134fccd.png)

(1)  
softmax 函数定义为：$\sigma(z)_i=\frac{e^{z_i}}{\sum_{j=1}^{K}e^{z_j}}$。  
对$z$求梯度要先求偏导数然后用联赛是法则求梯度。  
1. 对输入$z_i$的偏导
$$
\frac{\partial \sigma(z)_i}{\partial z_i} = \frac{\partial}{\partial z_i} \left( \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}} \right) = \frac{e^{z_i} \sum_{j=1}^{K} e^{z_j} - e^{z_i} e^{z_i}}{\left( \sum_{j=1}^{K} e^{z_j} \right)^2} = \sigma(z)_i (1 - \sigma(z)_i).
$$
   
2. 对输出$z_j$的偏导，$j \neq i$  
$$
\frac{\partial \sigma(z)_i}{\partial z_j} = \frac{\partial}{\partial z_j} \left( \frac{e^{z_i}}{\sum_{k=1}^{K} e^{z_k}} \right) = -\frac{e^{z_i} e^{z_j}}{\left( \sum_{k=1}^{K} e^{z_k} \right)^2} = -\sigma(z)_i \sigma(z)_j.
$$
  
通过链式法则
$$
\frac{\partial L}{\partial z_k} = \sum_{i=1}^{K} \frac{\partial L}{\partial \sigma(z)_i} \frac{\partial \sigma(z)_i}{\partial z_k}
$$  
写出矩阵形式为
$$
\nabla_z L = \begin{bmatrix}
\frac{\partial L}{\partial z_1} \\
\frac{\partial L}{\partial z_2} \\
\vdots \\
\frac{\partial L}{\partial z_K}
\end{bmatrix} = \begin{bmatrix}
\frac{\partial \sigma(z)_1}{\partial z_1} & \frac{\partial \sigma(z)_1}{\partial z_2} & \cdots & \frac{\partial \sigma(z)_1}{\partial z_K} \\
\frac{\partial \sigma(z)_2}{\partial z_1} & \frac{\partial \sigma(z)_2}{\partial z_2} & \cdots & \frac{\partial \sigma(z)_2}{\partial z_K} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial \sigma(z)_K}{\partial z_1} & \frac{\partial \sigma(z)_K}{\partial z_2} & \cdots & \frac{\partial \sigma(z)_K}{\partial z_K}
\end{bmatrix} \begin{bmatrix}
\frac{\partial L}{\partial \sigma(z)_1} \\
\frac{\partial L}{\partial \sigma(z)_2} \\
\vdots \\
\frac{\partial L}{\partial \sigma(z)_K}
\end{bmatrix}
$$

因此
$$
\nabla_z \sigma(z) = \begin{bmatrix}
\frac{\partial \sigma(z)_1}{\partial z_1} & \frac{\partial \sigma(z)_1}{\partial z_2} & \cdots & \frac{\partial \sigma(z)_1}{\partial z_K} \\
\frac{\partial \sigma(z)_2}{\partial z_1} & \frac{\partial \sigma(z)_2}{\partial z_2} & \cdots & \frac{\partial \sigma(z)_2}{\partial z_K} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial \sigma(z)_K}{\partial z_1} & \frac{\partial \sigma(z)_K}{\partial z_2} & \cdots & \frac{\partial \sigma(z)_K}{\partial z_K}
\end{bmatrix} = \begin{bmatrix}
\sigma(z)_1 (1 - \sigma(z)_1) & -\sigma(z)_1 \sigma(z)_2 & \cdots & -\sigma(z)_1 \sigma(z)_K \\
-\sigma(z)_2 \sigma(z)_1 & \sigma(z)_2 (1 - \sigma(z)_2) & \cdots & -\sigma(z)_2 \sigma(z)_K \\
\vdots & \vdots & \ddots & \vdots \\
-\sigma(z)_K \sigma(z)_1 & -\sigma(z)_K \sigma(z)_2 & \cdots & \sigma(z)_K (1 - \sigma(z)_K)
\end{bmatrix}.
$$  
可以看出，Softmax 函数对于 $z$ 的梯度是一个 $K \times K$ 的矩阵，其对角线元素为 $\sigma(z)_i (1 - \sigma(z)_i)$，非对角线元素为 $-\sigma(z)_i \sigma(z)_j$。这个矩阵也被称为 Softmax 函数的 Jacobian 矩阵。

(2)  
MLE 和交叉熵得到的损失函数是相同的。  
我们来推导一下为什么 MLE 和交叉熵在 $K$ 分类问题中是等价的。

首先，我们来看看 MLE 的损失函数。假设我们有 $N$ 个训练样本，每个样本的真实标签为 $y_i \in \{0, 1\}^K$，其中 $y_{ik} = 1$ 表示第 $i$ 个样本属于第 $k$ 个类，否则为 $0$。那么，MLE 的损失函数为：

$$
L_{\text{MLE}} = - \sum_{i=1}^{N} \log P(y_i|z_i)
$$

其中，$P(y_i|z_i)$ 表示模型对于第 $i$ 个样本的条件概率，即模型认为第 $i$ 个样本属于每个类的概率。由于我们假设模型的输出为 softmax 函数的输出，因此 $P(y_i|z_i)$ 可以表示为：

$$
P(y_i|z_i) = \prod_{k=1}^{K} \sigma(z_i)_k^{y_{ik}}
$$

其中，$\sigma(z_i)_k$ 表示模型对于第 $i$ 个样本属于第 $k$ 个类的预测概率。将上式代入到 $L_{\text{MLE}}$ 中，可得：

$$
L_{\text{MLE}} = - \sum_{i=1}^{N} \sum_{k=1}^{K} y_{ik} \log \sigma(z_i)_k
$$

接下来，我们来看看交叉熵的损失函数。假设我们有 $N$ 个训练样本，每个样本的真实标签为 $y_i \in \{0, 1\}^K$，那么交叉熵的损失函数为：

$$
L_{\text{CE}} = - \sum_{i=1}^{N} \sum_{k=1}^{K} y_{ik} \log \hat{y}_{ik}
$$

其中，$\hat{y}_{ik}$ 表示模型对于第 $i$ 个样本属于第 $k$ 个类的预测概率。由于我们假设模型的输出为 softmax 函数的输出，因此 $\hat{y}_{ik} = \sigma(z_i)_k$，因此可以将上式写成：

$$
L_{\text{CE}} = - \sum_{i=1}^{N} \sum_{k=1}^{K} y_{ik} \log \sigma(z_i)_k
$$

可以看出，MLE 和交叉熵的损失函数是相同的。这是因为 softmax 函数的输出是一个概率分布，它满足 $\sum_{k=1}^{K} \sigma(z)_k = 1$，因此 MLE 和交叉熵都是在最大化对数似然函数，即最大化 $\sum_{i=1}^{N} \sum_{k=1}^{K} y_{ik} \log(\sigma(z_i)_k)$。因此，MLE 和交叉熵在 $K$ 分类问题中是等价的。

需要注意的是，MLE 和交叉熵在二分类问题中也是等价的，因为二分类问题可以看作是 $K = 2$ 的特殊情况。在二分类问题中，MLE 和交叉熵的损失函数可以简化为：

$$
L_{\text{MLE}} = L_{\text{CE}} = - \sum_{i=1}^{N} y_i \log \hat{y}_i + (1 - y_i) \log (1 - \hat{y}_i)
$$

其中，$y_i \in \{0, 1\}$ 是第 $i$ 个样本的真实标签，$\hat{y}_i \in [0, 1]$ 是模型对于第 $i$ 个样本属于正类的预测概率。

前两问得到的结果就是 softmax 函数求梯度的方法和损失函数

```python
import numpy as np

# 设置每个类别的数据数量
num_samples_per_class = 100

# 设置每类的平均值和标准偏差 
means = [(2, -2), (2, 2), (-2, -2), (-2, 2)]
stds = [0.8, 0.8, 0.8, 0.8]

# 生成每个类的数据
data = []
for i in range(4):
    x = np.random.normal(means[i][0], stds[i], num_samples_per_class)
    y = np.random.normal(means[i][1], stds[i], num_samples_per_class)

    # 添加噪声
    num_noise = 10
    noise_x = np.random.normal(means[(i+1)%4][0], stds[(i+1)%4], num_noise)
    noise_y = np.random.normal(means[(i+1)%4][1], stds[(i+1)%4], num_noise)
    x = np.concatenate((x, noise_x))
    y = np.concatenate((y, noise_y))
    labels = np.concatenate((np.full(num_samples_per_class, i), np.full(num_noise, (i+1)%4)))

    data.append(np.column_stack((x, y, labels)))

# 连接数据
data = np.concatenate(data)

# 打乱数据
np.random.shuffle(data)

# 保存
np.savetxt('multiclass.csv', data, delimiter=',', fmt='%.3f')
```

```python
import numpy as np

def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def cross_entropy_loss(y_true, y_pred):
    return -np.sum(y_true * np.log(y_pred), axis=1)

def gradient_descent(X, y, learning_rate, num_iterations):
    num_samples, num_features = X.shape
    num_classes = np.unique(np.argmax(y, axis=1)).size

    # 初始化参数
    W = np.random.randn(num_features, num_classes)
    b = np.zeros(num_classes)

    for i in range(num_iterations):
        # 前向传播
        z = X.dot(W) + b
        y_pred = softmax(z)

        # 计算损失
        loss = np.mean(cross_entropy_loss(y, y_pred))
        print("Iteration %d, Loss: %.4f" % (i+1, loss))

        # 反向传播
        dz = y_pred - y
        dW = X.T.dot(dz)
        db = np.sum(dz, axis=0)

        # 更新参数
        W -= learning_rate * dW
        b -= learning_rate * db

    return W, b

# 加载数据集
data = np.loadtxt('multiclass.csv', delimiter=',')
X = data[:, :2]
y = data[:, 2]
y = np.eye(4)[y.astype(int)]

# 训练模型
learning_rate = 0.1
num_iterations = 1000
W, b = gradient_descent(X, y, learning_rate, num_iterations)

# 测试模型
def predict(X, W, b):
    z = X.dot(W) + b
    y_pred = softmax(z)
    return np.argmax(y_pred, axis=1)

test_data = np.array([[2, 2], [-2, 2], [2, -2]])
predictions = predict(test_data, W, b)
print("Predictions:", predictions)
```

10. 关于随机变量$X$存在两个概率分布$p(x)$和$q(x)$，证明其 KL 散度总为非负。即$D_{KL}(p\parallel q) \geq 0$，并且当且仅当$p(x)=q(x)$时，$D_{KL}(p\parallel q)=0$。

答：
- 非负性  
  - 凸函数：对于任意的$t_1,t_2>0$和$\lambda \in [0,1]$，都有$f(\lambda t_1 +(1-\lambda)t_2) \leq \lambda f(t_1)+(1-\lambda)f(t_2)$，即为任意两点的连线都在函数之上（和一般的凹凸认识不同，建议深入学习）  
  - Jensen 不等式：若$f(x)$是凸函数，$f(a_1x_1+a_2x_2+...+a_nx_n) \leq a_1f(x_1)+a_2f(x_2)+...+a_nf(x_n)$不等式成立  
  - $-\log(x)是凸函数。KL 散度结合 jensen 不等式：$$D_{KL}=\sum_{i=1}^{n}P(x_i) \times (-\log(\frac{Q(x_i)}{P(x_i)})) \ geq -\log(\sum^{n}_{i=1}P(x_i)\times \frac{Q(x_i)}{P(x_i)})=-\log(\sum^{n}_{i=1}Q(x_i))$$  $$\sum^{n}_{i=1}Q(x_i)=1$$  
    可得$$D_{KL} \geq -\log(1)=0$$

- 等于零
  - $p(x) > 0$
  - 当且仅当$p(x)=q(x)才能使得\log(\frac{p(x)}{q(x)}=1)=0$，才能使得$D_{KL}(p||q)=0$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

x = np.linspace(-5, 5, 100)
y = f(x)

x1 = -2
x2 = 3
y1 = f(x1)
y2 = f(x2)

k = (y2 - y1) / (x2 - x1)
b = y1 - k * x1

plt.plot(x, y, label='$f(x) = x^2$')
plt.plot([x1, x2], [y1, y2], 'r--', label='Secant line')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.show()

```
